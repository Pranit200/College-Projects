## Hypothesis Testing
**Using Chi-Square Testing**

- Is there a relationship between occupation and gender?

```python
df1.groupby(['gender','occupation']).count()['age']
```

```python
crtab = pd.crosstab(df1['occupation'],df1['gender'])
crtab
```

```python
from scipy import stats
test_statistic,p,dof,_=stats.chi2_contingency(crtab)
print(test_statistic)
print(p)
```

```python
if p >0.05:
 print('There is no relation between the two variable')
else:
 print('There is some significance relation between the two variables')
```

## Now we will handle imbalanced data

```python
from imblearn.over_sampling import SMOTE, RandomOverSampler
# apply the SMOTE oversampling technique to the training data
undersample = SMOTE()
X_train_oversampled, X_test_oversampled = undersample.fit_resample(X_train, X_test)
Y_valid_oversampled,Y_test_oversampled=undersample.fit_resample(Y_valid,Y_test)
Z_train_oversampled,Z_test_oversampled=undersample.fit_resample(x,Y)

print('Size of training data in oversampling')
print(X_train_oversampled.shape,X_test_oversampled.shape)
print(Y_valid_oversampled.shape,Y_test_oversampled.shape)
print(Z_train_oversampled.shape,Z_test_oversampled.shape)
```

```python
print(Z_test_oversampled.value_counts(normalize=True))
```

```python
# Calculate odds ratio
positive_class_index = 1  # Assuming positive class is labeled as 1
negative_class_index = 0  # Assuming negative class is labeled as 0
class_probabilities = adboost.predict_proba(Z_train_oversampled)
positive_probabilities = class_probabilities[:, positive_class_index]
negative_probabilities = class_probabilities[:, negative_class_index]
odds_ratio = positive_probabilities / negative_probabilities
print("Odds Ratio:", odds_ratio)
```

### The odds ratio represents the ratio of the predicted probabilities of the positive class (label 1) to the predicted probabilities of the negative class (label 0) for each instance in the test set.
- Odds Ratio < 1:
> The model is more confident in classifying these instances as the negative class.
- Odds Ratio > 1:
> This means that the model is more confident in classifying these instances as the positive class.
- Odds Ratio = 1:
> The model does not favor one class over the other and considers them equally likely.
